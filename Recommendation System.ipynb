```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv(r"C:\Users\amish\Downloads\archive (5)\25k IMDb movie Dataset.csv")
```

```python
data.head()
```

```python
data.rename(columns = {'Plot Kyeword': 'Plot_Keyword','Top 5 Casts':'Top_5_Casts','movie title':'Title'},inplace = True)
```

```python
data.head()
```

```python
features = ['Rating','Generes','Plot_Keyword','Director','Top_5_Casts']
```

```python
def combine_features(row):
 return row['Rating']+' '+row['Generes']+' '+row['Plot_Keyword']+' '+row['Director']+' '+row['Top_5_Casts']
```

```python
#filling all NaNs with blank string
for feature in features:
    data[feature] = data[feature].fillna(' ') 
```

```python
 #applying combined_features() method over each rows of dataframe and storing the combined string in “combined_features” column
data['combine_features'] = data.apply(combine_features,axis=1)
```

```python
data.rename_axis('index')
```

```python
data.head()
```

```python
cv = CountVectorizer() #creating new CountVectorizer() object
count_matrix = cv.fit_transform(data['combine_features']) #feeding combined strings(movie contents) to CountVectorizer() object
```

```python
count_matrix
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
def get_title_from_index(index):
    return data[data.index == index]['Title'].values[0]
def get_index_from_title(Title):
    return data.index[data['Title'] == Title][0]
```

```python
movie_user_likes = 'The Nun'
movie_index = get_index_from_title(movie_user_likes)
similar_movies = list(enumerate(cosine_sim[movie_index]))
#accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
print("Top 5 similar movies to "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```
